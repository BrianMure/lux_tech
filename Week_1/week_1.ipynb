```python
#import libraries
import numpy as np
import pandas as pd
```

### 1.0 Data Collection

```python
#read csv file
df = pd.read_csv("./1. Weather Data.csv")

#display first 10 rows
df.head(10)
```

```python
#display last 10 rows
df.tail(10)
```

### 2.0 Shape of the Dataset

```python
#get the shape
print(f"The dataframe has {df.shape[0]} rows and {df.shape[1]} columns")
```

### 3.0 Find Records where Weather Is Clear

```python
#get weather is clear
df[df['Weather'] == 'Clear']
```

- The records are 1326 where the weather was clear

### 4.0 Count Number of Times wind speed was exactly 4km/hr

```python
#loc wind speed
wind_speed = df[df['Wind Speed_km/h'] == 4]
print(f"The number of times wind speed was exactly 4km/hr is {wind_speed.shape[0]} times")
```

### 5.0 Check Null values

```python
#check for null values
df.isnull().sum()
```

### 6.0 Rename Weather column

```python
#rename column
df = df.rename(columns={'Weather': 'Weather_Condition'})

#check if it has changed
df.columns
```

### 7.0 Check Mean Visibility


```python
#use mean function
df['Visibility_km'].mean()

print(f"The mean visibility is {round(df['Visibility_km'].mean(), 3)}")
```

### 8.0 Wind Speed and Visibility

```python
#check for records where wind speed is greater than 24km/hr
wind_speed_and_visibilty = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)]
wind_speed_and_visibilty.shape[0]
print(f"The number of records are {wind_speed_and_visibilty.shape[0]}")
```

### 9.0 Mean Value for Each Weather Condition

```python
#mean value for each weather condition
weather_count = df['Weather_Condition'].value_counts()
weather_mean = weather_count.mean()
weather_mean
```

### 10. Instances of Clear Weather and Relative Humidity and Visibility

```python
#check
clear_weather_and_visibilty = df[(df['Weather_Condition'] > 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40) ]
clear_weather_and_visibilty
```

### 11. Number of Weather Conditions that include snow

```python
#number that include snow
snow_count = df[df['Weather_Condition'] == 'Snow']
print(f"The number of weather conditions that include snow are {snow_count.shape[0]}.")
```
